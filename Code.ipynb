### Reverse String

```python
input_str=str(input('Enter the string: '))
rev_str=''

for i in input_str:
    rev_str = i+ rev_str
print(rev_str)    
```

### Check Palindrome

```python
## Str Input


input_str = input('Enter the string: ')
pal_str = ''

for i in input_str:
    pal_str = i + pal_str

if pal_str == input_str:
    print('String is palindrome')
else:
    print('String is not palindrome')

```

```python
##  Int Input


num = int(input('Enter a number: '))
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if temp == rev:
    print('Number is palindrome')
else:
    print('Number is not palindrome')


```

### Find Duplicate Elements 

```python
num = [3,4,5,7,3]

tem = []

for i in num:
    if i in tem:
        print('Duplicate found:', i)
        break
    tem.append(i)
else:
    print('Not found')
```

### Remove Duplicates 

```python
num = [3,4,5,7,3]

tem = []

for i in num:
    if i in tem:
       break
    else:
        tem.append(i)    
print(tem)        
   
```

### Second Largest Number 

```python
num = [10, 20, 4, 45, 99]

temp = 0

for i in num:
    if i > temp:
        temp = i

num.remove(temp)

second = 0

for i in num:
    if i > second:
        second = i

print('Second largest number is:', second)
```

### Count Even and Odd 

```python
num = [2,5,7,8,3,4]

even = 0
odd = 0

for i in num:
    if i % 2 == 0:
        even = even + 1
    else:
        odd = odd + 1

print('Even:', even)
print('Odd:', odd)         
```

###  First Non-Repeating Character 

```python
s = "aabbcde" 
 
for i in range(len(s)): 
    count = 0 
    for j in range(len(s)): 
        if s[i] == s[j]: 
            count += 1 
    if count == 1: 
        print(s[i]) 
        break 
```

### Factorial

```python
n = 5
fact = 1

if n == 0 or n == 1:
    fact = 1
else:
    for i in range(1, n + 1):
        fact *= i

print(fact)  
```

### Fibonacci Series 

```python
n = 5 
a, b = 0, 1 
 
for i in range(n): 
    print(a) 
    a, b = b, a + b 
```

###  Check Prime Number 

```python
num = 6

for i in range(2, num // 2 + 1):
    if num % i == 0:
        print('Not prime')
        break
else:
    print('Prime')
```

###  Sum of Digits 

```python
no = int(input('Enter the No: '))
temp = 0

while no > 0:
    temp += no % 10
    no = no // 10

print(temp)
```

### frequency of Elements 

```python
list_no = [2,3,4,3,5,2]
temp = []

for i in list_no:
    if i not in temp:
        temp.append(i)
        print(f'Count of {i}: {list_no.count(i)}') 
    
```

```python
list_no = [2,3,4,3,5,2,3]
temp = []

for i in list_no:
    if i not in temp:
        temp.append(i)

        count = 0
        for j in list_no:
            if i == j:
                count += 1

        print(f'Count of {i}: {count}')
```

### Most Frequent Element

```python
list_no = [2,3,4,3,5,2,3]

max_count = 0
most_freq = 0

for i in list_no:
    count = 0

    for j in list_no:
        if i == j:
            count += 1

    if count > max_count:
        max_count = count
        most_freq = i

print('Most frequent element:', most_freq)
```
