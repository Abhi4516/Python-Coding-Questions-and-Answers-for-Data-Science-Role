### Reverse String

```python
input_str=str(input('Enter the string: '))
rev_str=''

for i in input_str:
    rev_str = i+ rev_str
print(rev_str)    
```

### Check Palindrome

```python
## Str Input


input_str = input('Enter the string: ')
pal_str = ''

for i in input_str:
    pal_str = i + pal_str

if pal_str == input_str:
    print('String is palindrome')
else:
    print('String is not palindrome')

```

```python
##  Int Input


num = int(input('Enter a number: '))
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if temp == rev:
    print('Number is palindrome')
else:
    print('Number is not palindrome')


```

### Find Duplicate Elements 

```python
num = [3,4,5,7,3]

tem = []

for i in num:
    if i in tem:
        print('Duplicate found:', i)
        break
    tem.append(i)
else:
    print('Not found')
```

### Remove Duplicates 

```python
num = [3,4,5,7,3]

tem = []

for i in num:
    if i in tem:
       break
    else:
        tem.append(i)    
print(tem)        
   
```

### Second Largest Number 

```python
num = [10, 20, 4, 45, 99]

temp = 0

for i in num:
    if i > temp:
        temp = i

num.remove(temp)

second = 0

for i in num:
    if i > second:
        second = i

print('Second largest number is:', second)
```
